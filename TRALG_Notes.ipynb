```python
# AAPL close prices in time window in 2d array 
close_AAPL = close_price.AAPL.copy().to_frame()
close_AAPL
```

```python
# Calc Increase or Decrease form prev day
close_AAPL["PrevDay"]=close_AAPL.shift()
```

```python
# change in price from yesterday
close_AAPL["Diff"]=close_AAPL.AAPL.sub(close_AAPL.PrevDay)
close_AAPL
```

```python
close_AAPL["% Change"] = close_AAPL.Diff.div(close_AAPL.PrevDay).mul(100)
close_AAPL
```

```python
# New Techniques to find Diff and $ Change
close_AAPL["Diff 2"]=close_AAPL.AAPL.diff(periods=1)
close_AAPL["% Change 2"] = close_AAPL.AAPL.pct_change(periods=1).mul(100)
close_AAPL
```

# Mean, Variance, Standard Deviation

```python
dReturn = close_AAPL.pct_change().dropna()
dReturn
```

```python
dReturn.plot(kind="hist",figsize=(12,8), bins=100)
plt.show()
```

```python
# average % daily change
daily_mean = dReturn.mean()
daily_mean
```

```python
# calc variance
daily_var=dReturn.var()
daily_var
```

```python
# calc std deviation
daily_std = np.sqrt(daily_var)
daily_std
```

```python
# annual mean return
annual_mean = daily_mean*252
annual_mean
```

```python
# annual variance
annual_var = daily_var*252
annual_var
```

```python
# Annual Std Deviation
annual_std = np.sqrt(annual_var)
annual_std
```

```python
ret.describe().T
```

```python
# NOTE .loc(rows, columns)
summary = ret.describe().T.loc[:,["mean","std"]]
summary
```

```python
#
summary["mean"] = summary["mean"]*252
summary["std"] = summary["std"]*np.sqrt(252)
summary
```

```python
summary.plot.scatter(x="std",y="mean",figsize=(12,8),s=50,fontsize=15)
for i in summary.index:
    plt.annotate(i,xy=(summary.loc[i,"std"]+0.002,summary.loc[i,"mean"]+0.002),size=15)
plt.xlabel("Annual Risk ( YRLY STD )")
plt.ylabel("Annual Return ( YRLY Mean )")
plt.title("Risk/Return",fontsize=25)
plt.show()
```

It is important in finacial strategies to properly analyze risk. This chart shows if pinned with the choice to buy MSFT AAPL or META, MSFT is the clear right choice for it has the lowest risk (std) while maintaining a relatively high return

# Simple and Log Returns

```python
df = pd.DataFrame(index=[2016,2017,2018],data=[100,50,95],columns=["Price"])
df
```

```python
simpleret=df.pct_change().dropna()
simpleret
```

```python
simRetMean = simpleret.mean()
simRetMean
```

```python
# Lets Calculate what the 2018 price will be based on simple mean return
strt = df.loc[2016,"Price"]
simRetMeanF = strt*(simRetMean+1)*(simRetMean+1)
simRetMeanF
```

This displays how mean returns can be misleading because the way they are calculated. In this case, the stock progressed from 100% -> 50% -> 190%. We can see that the stock grew rapidly after losing a significant amount of value. Therfore the mean will factor the recovery/bounce into its result. 

Lets use Log Returns

```python
logret = np.log(df/df.shift()).dropna()
logret
```

```python
logRetMean = logret.mean()
```

```python
logRetMeanF = strt*np.exp(logRetMean*2)
logRetMeanF
```

Accurate Results are given by Log Mean Return
